```python
# Data Structures in Python

# Topic: List
print("### List Exercises ###")
# Q1: Create a list of 5 random numbers and print the list.
import random
my_list = [random.randint(1, 100) for _ in range(5)]
print("Original List:", my_list)

# Q2: Insert 3 new values to the list and print the updated list.
new_values = [200, 300, 400]
my_list.extend(new_values)
print("Updated List:", my_list)

# Q3: Use a for loop to print each element in the list.
print("Elements in the List:")
for element in my_list:
    print(element)

print("\n")  # Spacer for better readability

# Topic: Dictionary
print("### Dictionary Exercises ###")
# Q1: Create a dictionary with keys 'name', 'age', and 'address' and values 'John', 25, and 'New York'.
my_dict = {'name': 'John', 'age': 25, 'address': 'New York'}
print("Original Dictionary:", my_dict)

# Q2: Add a new key-value pair to the dictionary with key 'phone' and value '1234567890'.
my_dict['phone'] = '1234567890'
print("Updated Dictionary:", my_dict)

print("\n")  # Spacer for better readability

# Topic: Set
print("### Set Exercises ###")
# Q1: Create a set with values 1, 2, 3, 4, and 5.
my_set = {1, 2, 3, 4, 5}
print("Original Set:", my_set)

# Q2: Add the value 6 to the set.
my_set.add(6)
print("Set after Adding 6:", my_set)

# Q3: Remove the value 3 from the set.
my_set.discard(3)  # Using discard to avoid error if the element doesn't exist.
print("Set after Removing 3:", my_set)

print("\n")  # Spacer for better readability

# Topic: Tuple
print("### Tuple Exercises ###")
# Q1: Create a tuple with values 1, 2, 3, and 4.
my_tuple = (1, 2, 3, 4)
print("Tuple:", my_tuple)

# Q2: Print the length of the tuple.
print("Length of the Tuple:", len(my_tuple))
```
